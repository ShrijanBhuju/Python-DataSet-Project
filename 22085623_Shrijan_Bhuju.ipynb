# 2. Data Preparation

## Question 1: Write a python program to load data into pandas DataFrame

```python
import pandas as pd #imports pandas.
import numpy as np #imports numpy.

salaryDataFrame = pd.read_csv("DataScienceSalaries.csv") # reads the uploaded .csv file.
print(salaryDataFrame)
```

## Question 2: Write a python program to remove unnecessary columns i.e., salary and salary currency.

```python
drpCol = {"salary", "salary_currency"} #Assigning column names to drpCol variable
salaryDataFrame = salaryDataFrame.drop(columns = drpCol) #Drops columns assigned to drpCol
print(salaryDataFrame)
```

## Question 3: Write a python program to remove the NaN missing values from updated dataframe.

```python
salaryDataFrame.notnull() #Checks if the updated dataframe has missing values.
```

```python
salaryDataFrame = salaryDataFrame.dropna() #removes missing values from recent dataframe
```

## Question 4: Write a python program to check duplicates value in the dataframe.

```python
salaryDataFrame.duplicated() #Checks for a duplicated value in the dataframe
```

## Question 5: Write a python program to see the unique values from all the columns in the dataframe.

```python
for column in salaryDataFrame.columns: #For Loop
  uniqueVal = salaryDataFrame[column].unique() #to view unique value from all column and assign to uniqueVal
  print("Unique values in column", column)
  print(uniqueVal)
```

## Question 6: Rename the experience level columns as below.
SE – Senior Level/Expert
MI – Medium Level/Intermediate
EN – Entry Level
EX – Executive Level

```python
salaryDataFrame['experience_level'] = salaryDataFrame['experience_level'].replace(
    {'SE': 'Senior Level/Expert',
     'MI': 'Medium Level/Intermediate',
     'EN':'Entry Level',
     'EX': 'Executive Level'})
print(salaryDataFrame)
```

# 3. Data Analysis

## Question 1: Write a Python program to show summary statistics of sum, mean, standard deviation, skewness, and kurtosis of any chosen variable.

```python
chosenVariable = salaryDataFrame.salary_in_usd
sum = chosenVariable.sum() #sum calculation
mean = chosenVariable.mean() #mean Calculation
standardDeviation = chosenVariable.std() #standard Deviation Calculation
skewness = chosenVariable.skew() #Skewness calculation
kurtosis = chosenVariable.kurtosis() #Kurtosis Calculation

#printing the above calculations
print("Sum is: ", sum)
print("Mean is: ", mean)
print("Standand Deviation is: ", standardDeviation)
print("Skewness is: ", skewness)
print("Kurtosis is: ", kurtosis)
```

## Question 2: Write a Python program to calculate and show correlation of all variables.

```python
allVariable = ['work_year', 'salary_in_usd', 'remote_ratio'] #all variable = all int/float type variables.
correlation = salaryDataFrame[allVariable].corr() #Correlation calaculation
correlation # prints correlation calculations
```

# 4. Data Exploration

## Question 1: Write a python program to find out top 15 jobs. Make a bar graph of sales as well.

```python
jobs=salaryDataFrame["job_title"].value_counts().head(15) #Data list of job_title column in salaryDataFrame
print(jobs)
```

```python
import matplotlib.pyplot as plt #importing matplot library as plt
plt.figure(figsize=(15,6)) #Figure Size
plt.bar(jobs.index, jobs.values, color='blue') #bar with blue color
plt.title('top 15 jobs') # title of the plot
plt.xlabel('job titles') #label on x-axis
plt.ylabel('number of jobs') #Label on y-axis
plt.xticks(rotation=90)
plt.show() #bar graph printout
```

## Question 2: Which job has the highest salaries? Illustrate with bar graph.

```python
highestPaidJob = salaryDataFrame[salaryDataFrame['salary_in_usd'] == salaryDataFrame['salary_in_usd'].max()]['job_title'].tolist()[0] #tolist = converting numPy array into a python list
highestSalaary = salaryDataFrame[salaryDataFrame['salary_in_usd'] == salaryDataFrame['salary_in_usd'].max()]['salary_in_usd'].tolist()[0]
plt.figure(figsize=(20, 10))
plt.bar(salaryDataFrame['job_title'], salaryDataFrame['salary_in_usd'], color='blue')
plt.xlabel('Job Title')
plt.ylabel('Salary')
plt.title('Salaries for Different Jobs')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

## Question 3: Write a python program to find out salaries based on experience level. Illustrate it through bar graph

```python
import seaborn as sns #impports seaborn library
```

```python
avg_salary_by_experience = (
    salaryDataFrame.groupby('experience_level')['salary_in_usd'].mean().sort_values(ascending=False)
)
plt.figure(figsize=(8, 6))
sns.barplot(x=avg_salary_by_experience.index, y=avg_salary_by_experience.values, color='blue')
plt.title('Average Salary by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Average Salary (USD)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

## Question 4: Write a Python program to show histogram and box plot of any chosen different variables. Use proper labels in the graph.

```python
# To show Histogram
plt.figure(figsize=(12, 10))
plt.hist(
    salaryDataFrame['salary_in_usd'],
    bins=20,
    color='blue',
    edgecolor='black',
    linewidth=1
)
plt.title('Histogram of salary_in_usd')
plt.xlabel('salary_in_usd')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#To show box plot
plt.figure(figsize=(8, 6))
sns.boxplot(
    x='experience_level',
    y='salary_in_usd',
    showmeans=True,
    data=salaryDataFrame,
    linewidth=10
)
plt.title('Box Plot of salary_in_usd by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('salary_in_usd')
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()
plt.show()
```
